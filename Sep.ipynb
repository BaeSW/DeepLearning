```python
import csv
import shutil

 
f = open('train.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
for (num, count) in rdr:
    if(int(num)<10001):
        continue
    shutil.copy("valid\\valid_image\\P_"+str(num)+".bmp" , "valid\\"+str(count)+"\\P_"+str(num)+".bmp")
    print('진행률 : ', num)
    
f.close() 
```
